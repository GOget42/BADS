```python
import pandas as pd
```

## Add a Date Column

```python
flights['DATE'] = pd.to_datetime(flights[['YEAR', 'MONTH', 'DAY']].astype(str).agg('-'.join, axis=1)).dt.strftime('%d.%m.%Y')
flights.head()
```

```python
flights.columns
```

## Plotten der Häufigkeit der Dateneinträge nach Abflug- und Destinationsflughafen

```python
import matplotlib.pyplot as plt

# Häufigkeit der Dateneinträge nach ORIGIN_AIRPORT
origin_counts = flights['ORIGIN_AIRPORT'].value_counts()

# Häufigkeit der Dateneinträge nach DESTINATION_AIRPORT
destination_counts = flights['DESTINATION_AIRPORT'].value_counts()

# Plotten der Häufigkeit nach ORIGIN_AIRPORT
plt.figure(figsize=(15, 7))
origin_counts.plot(kind='bar')
plt.title('Häufigkeit der Dateneinträge nach ORIGIN_AIRPORT')
plt.xlabel('ORIGIN_AIRPORT')
plt.ylabel('Häufigkeit')
plt.show()

# Plotten der Häufigkeit nach DESTINATION_AIRPORT
plt.figure(figsize=(15, 7))
destination_counts.plot(kind='bar')
plt.title('Häufigkeit der Dateneinträge nach DESTINATION_AIRPORT')
plt.xlabel('DESTINATION_AIRPORT')
plt.ylabel('Häufigkeit')
plt.show()
```

## Die 15 häufigsten Abflug- und Destinationsflughäfen

```python
# Plotten der 5 häufigsten Dateneinträge nach ORIGIN_AIRPORT
plt.figure(figsize=(10, 5))
origin_counts.head(15).plot(kind='bar')
plt.title('Top 5 häufigste Dateneinträge nach ORIGIN_AIRPORT')
plt.xlabel('ORIGIN_AIRPORT')
plt.ylabel('Häufigkeit')
plt.show()

# Plotten der 5 häufigsten Dateneinträge nach DESTINATION_AIRPORT
plt.figure(figsize=(10, 5))
destination_counts.head(15).plot(kind='bar')
plt.title('Top 5 häufigste Dateneinträge nach DESTINATION_AIRPORT')
plt.xlabel('DESTINATION_AIRPORT')
plt.ylabel('Häufigkeit')
plt.show()
```

## Konzentration auf die Flüge mit Wetterverspätung

```python

weather_delay_flights = flights[flights['WEATHER_DELAY'] > 0]
weather_delay_flights
```

```python
import matplotlib.pyplot as plt

# Häufigkeit der Dateneinträge nach ORIGIN_AIRPORT
origin_counts = weather_delay_flights['ORIGIN_AIRPORT'].value_counts()

# Häufigkeit der Dateneinträge nach DESTINATION_AIRPORT
destination_counts = weather_delay_flights['DESTINATION_AIRPORT'].value_counts()

# Plotten der 5 häufigsten Dateneinträge nach ORIGIN_AIRPORT
plt.figure(figsize=(10, 5))
origin_counts.head(15).plot(kind='bar')
plt.title('Top 5 häufigste Dateneinträge für Abflugflughäfen mit einer Verspätung aufgrund des Wetters')
plt.xlabel('Abflugflughafen')
plt.ylabel('Häufigkeit')
# Plotten der tatsächlichen Zahl der Häufigkeit in die Barchart für ORIGIN_AIRPORT
for i, v in enumerate(origin_counts.head(15)):
    plt.text(i, v + 10, str(v), ha='center', va='bottom')



# Plotten der 5 häufigsten Dateneinträge nach DESTINATION_AIRPORT
plt.figure(figsize=(10, 5))
destination_counts.head(15).plot(kind='bar')
plt.title('Top 5 häufigste Dateneinträge nach DESTINATION_AIRPORT')
plt.xlabel('DESTINATION_AIRPORT')
plt.ylabel('Häufigkeit')
plt.show()
```
